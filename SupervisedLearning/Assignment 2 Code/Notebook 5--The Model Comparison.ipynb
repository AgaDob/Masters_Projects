```python
import math
import numpy as np
import pandas as pd
import scipy.io as sio
import seaborn as sns
import matplotlib.pyplot as plt
import csv
import time
import collections

from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import polynomial_kernel
from numpy.linalg import inv, norm
from mpl_toolkits.mplot3d import axes3d
from numpy import mean
%matplotlib inline
```

### Test Error

```python
# Create lists for the plot
models = ['Polynomial 1-v-All', 'Gaussian 1-v-All']
x_pos = np.arange(len(models))
CTEs = [2.8, 2.5 ]
error = [0.3, 0.3]
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, CTEs, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10, color ='green')
ax.set_title('Comparison of Mean Test Errors using Optimum Parameters')
ax.set_xticks(x_pos)
ax.set_xticklabels(models)
ax.set_ylabel('% Test Error')
ax.yaxis.grid(False)

# Save the figure and show
plt.tight_layout()
plt.savefig('model_comparison.png')
plt.show()
```

### Time complexity

```python
# Create lists for the plot
models = ['Poly 1-v-All', 'RBF 1-v-All', 'Poly 1-v-1', 'KNN', 'SVM']
x_pos = np.arange(len(models))
CTEs = [22, 7 ,48, 19, 14 ]
error = [4, 1 ,12, 4, 1 ]
```
